```python
 def sentimentalAnalysis(data,columnname):
        
        #downloading vader_lexicon for the process
        import matplotlib.pyplot as plt
        import seaborn as sns
        "Importing Necessary Packeage"
        from nltk.sentiment.vader import SentimentIntensityAnalyzer
        sid = SentimentIntensityAnalyzer()
        
        "Deleting null pr empty value"
        data.dropna(inplace=True)
        
        "Checking for a comment"
        #sid.polarity_scores(data[columnname][93])
        
        "Creating respective columns"
        
        data['scores'] = data[columnname].apply(lambda commentText: sid.polarity_scores(commentText))
        data['compound']  = data['scores'].apply(lambda score_dict: score_dict['compound'])
        data['Negtive']  = data['scores'].apply(lambda score_dict: score_dict['neg'])
        data['Postive']  = data['scores'].apply(lambda score_dict: score_dict['pos'])
        data['Neutral']  = data['scores'].apply(lambda score_dict: score_dict['neu'])
        
        "Creating final pos or neg using compound score"
        data['comp_score'] = data['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
        plt.clf()
        #comp=sns.countplot(x = 'comp_score', hue = 'Name', data = data, palette = 'magma')
        #comp.figure.savefig("date_charts.png")
        "Checking how many pos and neg"
        posneg=pd.DataFrame(data['comp_score'].value_counts())
        return posneg,data
    
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#we perform data preprocessing with tfid
tfidf=TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')
dtm=tfidf.fit_transform(df["Chat"])
```

```python
#preprocessed data is fed into NMF
from sklearn.decomposition import NMF

#n is number of topics 
nmf_model=NMF(n_components=5,random_state=42)

#vectorized variable dtm is sent here as input; fit=only model has been created
nmf_model.fit(dtm)
```

```python
#for each of the 5 topics, we are displaying first 10 words for each topic 

for index, topic in enumerate(nmf_model.components_):
    results = ([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
    print(results)
```

```python
topic_results=nmf_model.transform(dtm)
df["Topic"]=topic_results.argmax(axis=1)
```

```python
# 1. Date-wise message count & graph to find who spoke the most
def plot_message_count(df):
    # Count messages per day for each sender
    date_wise_count = df.groupby(['Date', 'Name']).size().unstack(fill_value=0)
    
    # Plot the graph
    plt.figure(figsize=(10, 6))
    date_wise_count.plot(kind='bar')
    plt.title('Date-wise Message Count')
    plt.xlabel('Date')
    plt.ylabel('Number of Messages')
    plt.legend(title="Sender")
    plt.show()

plot_message_count(df)
```

```python
# 2. Flirt Encounter (Text input, Yes/No field for flirt detection, most & least talkative)

# Keywords to detect flirty messages
flirt_keywords = ['love', 'miss you', 'kiss', 'cute', 'babe']

def flirt_encounter(df, keywords):
    # Detect flirt messages using keywords
    df['Flirt'] = df['Chat'].apply(lambda x: 'Yes' if any(keyword in x.lower() for keyword in keywords) else 'No')

    # Group by 'Name' and check if anyone has flirted
    flirt_summary = df.groupby('Name')['Flirt'].apply(lambda x: 'Yes' if 'Yes' in x.values else 'No').reset_index()
    
    # Most & least talkative
    talkative_count = df['Name'].value_counts()
    most_talkative = talkative_count.idxmax()
    least_talkative = talkative_count.idxmin()
    
    # Print most and least talkative
    print("Most Talkative:", most_talkative)
    print("Least Talkative:", least_talkative)
    
    # Display flirt summary (Who flirted)
    display(flirt_summary)

# Run the flirt encounter function
flirt_encounter(df, flirt_keywords)
```

```python
# 3. Time Encounter (Most active date, day, hour and average number of messages)

def time_encounter(df):
    # Convert 'Date' to datetime
    df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%y', errors='coerce')
    
    # Convert 'Time' to datetime and extract the hour
    df['Time'] = pd.to_datetime(df['Time'], format='%I:%M %p', errors='coerce')
    df['Hour'] = df['Time'].dt.hour
    df['Day'] = df['Date'].dt.day_name()
    
    # Most active date, day, hour
    most_active_date = df['Date'].value_counts().idxmax()
    most_active_day = df['Day'].value_counts().idxmax()
    most_active_hour = df['Hour'].value_counts().idxmax()
    
    # Average number of messages
    avg_messages = df.groupby(['Date']).size().mean()

    print(f"Most Active Date: {most_active_date.date()}")
    print(f"Most Active Day: {most_active_day}")
    print(f"Most Active Hour: {most_active_hour}")
    print(f"Average Messages Per Day: {avg_messages:.2f}")

# Call the function
time_encounter(df)
```

```python
# 4. Media Encounter (Media Count, deleted message count, missed calls)
def media_encounter(df):
    # Counting media, deleted messages, missed calls
    media_count = df['Chat'].str.contains('<Media omitted>').sum()
    deleted_message_count = df['Chat'].str.contains('This message was deleted').sum()
    missed_voice_call = df['Chat'].str.contains('Missed voice call').sum()
    missed_video_call = df['Chat'].str.contains('Missed video call').sum()
    
    print(f"Media Count: {media_count}")
    print(f"Deleted Messages: {deleted_message_count}")
    print(f"Missed Voice Calls: {missed_voice_call}")
    print(f"Missed Video Calls: {missed_video_call}")

media_encounter(df)
```
